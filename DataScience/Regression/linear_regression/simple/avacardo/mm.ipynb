```python
def error(points, m, b):
    xs = points['volume']
    ys = points['price']
    error = 0
    for i, x in enumerate(xs):
        error += (ys[i] - (m*x+b)) ** 2
    return error/len(ys)
```

```python
def gradient(points, m, b, lr):
    xs = np.array(points['volume'])
    ys = np.array(points['price'])
    grad_m = 0
    grad_b = 0

    for i in range(len(xs)):
        grad_m += xs[i]*(ys[i] - (m*xs[i]+b))
        grad_b += (ys[i] - (m*xs[i]+b))
    grad_m = (-2/len(xs)*grad_m)
    grad_b = (-2/len(xs)*grad_b)
    
    m -= grad_m*lr
    b -= grad_b*lr
    return m, b
```

```python
def runner(points, m, b, itr, lr):
    for _ in range(itr):
        m, b = gradient(points, m, b, lr)
        print(error(data, m, b))
    return m, b
```

```python
m = 0
b = 0
print(error(data, m, b))
m, b = runner(data, m, b, 1000, 0.01)
print(error(data, m, b))
```
